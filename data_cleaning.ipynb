```python
import pandas as pd
import re
```

```python
df = pd.read_csv("C:/Users/filippovm/OneDrive - Milwaukee School of Engineering/Documents/CSC3320/Final_Project/GroceryDataset.csv")
df.dtypes
```

```python
def clean_rating(text)->float:
    if type(text) != float:
        result = re.search(r'\d+(\.?\d+)?', text)
        if result != None:
            return float(re.search(r'\d+(\.?\d+)?', text).group())
    return float('nan')

def clean_price(text)->float:
    text = re.sub(',', '', text)
    return float(text[1:])

def clean_discount(text)->float:
    if type(text) != float:
        result = re.search(r'\d+(\.?\d+)?', text)
        if result != None:
            return float(re.search(r'\d+(\.?\d+)?', text).group())
    return 0.0
```

```python
# clean ratings
df['Rating'] = df['Rating'].map(clean_rating)
```

```python
# NaN ratings ratio:
df['Rating'].isna().sum() / len(df)
```

```python
# drop NaN data that's not possible to use
indexes = df[(df['Price'].str.len() > 11) | (df['Price'].isna()) | (df['Currency'].isna())].index
df.drop(indexes, inplace=True)
len(df)
```

```python
# convert all prices to floats
df['Price'] = df['Price'].map(clean_price)
```

```python
# convert all discounts to floats
df['Discount'] = df['Discount'].map(clean_discount)
```

```python
# rename column name
df.rename(columns={'Sub Category':'Category'}, inplace=True)
```
